# Cleaning data

```python
df1 = df.copy(deep=True)

df1 = df1.drop(columns=['Farm Name','Lot Number','Mill','Uniformity',
                        'ICO Number','Company','Region',
                        'Producer','Bag Weight','In-Country Partner',
                        'Owner','Variety','Status',
                        'Clean Cup','Sweetness','Defects',
                        'Certification Body','Certification Address','Certification Contact'])

df1['Harvest Year'] = df1['Harvest Year'].apply(lambda x: int(x.split(' / ')[0]) if ' / ' in x else int(x))

df1['Altitude'] = df1['Altitude'].str.extract('(\d+)')[0]  # Extract numeric part and convert to integer
df1['Altitude'] = df1['Altitude'].astype(float).fillna(np.nan).astype('Int64') # Convert to integer, keeping NaN as NaN

df1['Processing Method'] = df1['Processing Method'].apply(lambda x: 
         'Anaerobic' if x in ['Anaerobico 1000h', 'Double Anaerobic Washed'] 
    else 'Honey'   if x in ['Honey,Mossto', 'Pulped natural / honey'] 
    else 'Natural' if x in ['Double Carbonic Maceration / Natural', 'Natural / Dry'] 
    else 'Washed'  if x in ['Semi Washed', 'Washed / Wet'] 
    else 'Others'
)


month_mapping = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}

df1[['G_MMM', 'G_YYYY']] = df1['Grading Date'].str.extract(r'(\w+) \w+, (\d+)')
df1['G_YYYY'] = df1['Grading Date'].str.split(',').str[1].str.strip().astype(int)
df1['G_MM'] = df1['G_MMM'].map(month_mapping)
df1['Grading Month'] = df1['G_YYYY'].astype(str) + df1['G_MM'].apply(lambda x: f'{x:02}') 
df1 = df1.drop(columns=['G_MMM','G_MM','G_YYYY'])

df1[['E_MMM', 'E_YYYY']] = df1['Expiration'].str.extract(r'(\w+) \w+, (\d+)')
df1['E_YYYY'] = df1['Expiration'].str.split(',').str[1].str.strip().astype(int)
df1['E_MM'] = df1['E_MMM'].map(month_mapping)
df1['Expire Month'] = df1['E_YYYY'].astype(str) + df1['E_MM'].apply(lambda x: f'{x:02}') 
df1 = df1.drop(columns=['E_MMM','E_MM','E_YYYY'])

df1 = df1.drop(columns=['Grading Date','Expiration'])

# print(df1['Harvest Year'].value_counts().sort_index())
# print(df1['Altitude'].value_counts().sort_index())
# print(df1['Processing Method'].value_counts().sort_index())
# print(df1['Grading Month'].value_counts().sort_index())
# print(df1['Expire Month'].value_counts().sort_index())
```

## Altitude Group

```python
# Create age groups using qcut
df1['Altitude_group'] = pd.qcut(df1['Altitude'], q=5, labels=False)

# Group by age_group to find min and max ages
Alt_summary = df1.groupby('Altitude_group').agg(
    min_Alt=('Altitude', 'min'),
    max_Alt=('Altitude', 'max')
).reset_index()

# Create age_range column in the summary DataFrame
# Alt_summary['Altitude_range'] = Alt_summary['min_Alt'].astype(str) + ' - ' + Alt_summary['max_Alt'].astype(str)
Alt_summary['Altitude_range'] = '<= ' + Alt_summary['max_Alt'].astype(str)

# Merge the age_range back to the original DataFrame
df1 = df1.merge(Alt_summary[['Altitude_group', 'Altitude_range']], on='Altitude_group', how='left')

df1[['Altitude', 'Altitude_group', 'Altitude_range']].head(10)
```

## Moisture Group

```python
df1['Moisture_group'] = pd.qcut(df1['Moisture Percentage'], q=5, labels=False)
df1['Moisture_group'].value_counts().sort_index()

result = df1.groupby('Moisture_group').agg(
    max_Moisture=('Moisture Percentage', 'max'),
    min_Moisture=('Moisture Percentage', 'min'),
    value=('Moisture_group', 'count')
).reset_index()

# result['Moisture_group'] = result['min_Moisture'].astype(str) + ' - ' + result['max_Moisture'].astype(str)
result['Moisture_group'] = '<= ' + result['max_Moisture'].astype(str)

result.head(10)
```

# Simple Statistics Analysis

## Cross Attributes

```python
columns_of_interest = ['Overall','Total Cup Points',
                       'Aroma','Flavor','Aftertaste','Acidity','Body','Balance']
attributes_of_interest = ['Country of Origin',
                          'Processing Method',
                          'Altitude_group',
                          'Moisture_group',
                          'Color'
                         ]

for col in columns_of_interest:
    print(f'==================== {col} ====================')    
    for att in attributes_of_interest:
        result = df1.groupby(att, as_index=False)[col].mean()
        result = result.sort_values(by=col, ascending=False)
        print(f"[{att} -- {col}]\n{result.head(5)}\n")
```
